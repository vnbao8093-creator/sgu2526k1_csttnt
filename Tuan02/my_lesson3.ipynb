#### 4. Sinh ngẫu nhiên dãy số
(a) Sinh ngẫu nhiên dãy số có 100 phần tử theo phân bố đều với giá trị trong đoạn [1, 10] và vẽ đồ thị tần số xuất hiện <br/> 
(b) Sinh ngẫu nhiên N diem (x, y) với y là hàm phân phối chuẩn $N(\mu = 1, \sigma^2 = 1.0)$: <br/>
 $$
 y = f\left(x| \mu, \sigma\right) = \frac{1}{\sigma\sqrt{2\pi}}exp\left(-\frac{\left(x - \mu\right)^2}{2\sigma^2}\right)
 $$
 nơi mà $x \in [\mu - 5\sigma, \mu + 5\sigma]$ <br/> 
 (c) Sinh ngẫu nhiên dãy số có 100000 phần tử theo phân bố chuẩn $N(\mu = 1, \sigma^2 = 1.0)$ và vẽ đồ thị tần số xuất hiện <br/> <br/> 

```python
import numpy as np
import matplotlib.pyplot as plt  # Khai báo thư viện pyplot trong matplotlib

data = np.random.randint(1, 11, size=10000)
value, cnt = np.unique(data, return_counts=True)
# Sử dụng hàm np.random.randint để sinh dãy số ngẫu nhiên
# Sử dụng hàm np.unique để trả về tần số xuất hiện
# ...

print("a. Sinh ngẫu nhiên theo phân bố đều")
print(f"+ Gia tri: {value}")
print(f"+ So lan xuat hien: {cnt}")
print("+ Do thi ham so xuat hien")
plt.figure(figsize=(6,6))      #  kich thuoc ban ve
plt.bar(value, cnt)            #  ve cac khoi chu nhat theo day x, y
plt.xlim(0, 11)                #  gioi han truc x
plt.ylim(0, np.max(cnt) + 10)  #  gioi han truc y
plt.savefig("4a.png")          #  luu do thi
plt.show()                     #  hien thi do thi
```

```python
# 1. Khai báo các tham số cho phân phối chuẩn N(mu=1, sigma^2=1)
mu = 1.0
sigma = 1.0

# 2. Xác định miền giá trị của x: [mu - 5*sigma, mu + 5*sigma]
x_min = mu - 5 * sigma
x_max = mu + 5 * sigma

# 3. Sinh ngẫu nhiên N điểm x trong khoảng [x_min, x_max]
# Công thức: x = rand * (b - a) + a
N = 100 # Bạn có thể điều chỉnh số lượng điểm để đồ thị đẹp hơn
x = np.random.rand(N) * (x_max - x_min) + x_min

# 4. Tính y theo hàm mật độ xác suất của phân phối chuẩn
tu_so = -(x - mu)**2
mau_so = 2 * sigma**2
y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(tu_so / mau_so)
# Sử dụm hàm np.random.rand() sinh số ngẫu nhiên trong [0,1) --> [0,1) * (b - a) + a --> [a, b)
# Sử dụng hàm sau để tính hàm xác suất chuẩn N(m, s): np.sqrt, np.exp
# ...

print("b. Sinh cặp (x, y) với y làm hàm phân phối chuẩn")
print(f"+ x: {x_min} <= min [{np.min(x): .2f}] <= max [{np.max(x): .2f}] <= {x_max}")
print(f"+ Ve ham so y(x)")
plt.figure(figsize=(6,6))      # kich thuoc ban ve
plt.scatter(x, y)              # ve cac diem (xi, yi)
plt.savefig("4b.png")          #  luu do thi
plt.show()                     #  hien thi do thi
```

```python
# 1. Khai báo các tham số
mu = 1.0
sigma = 1.0
N_points = 100000

# 2. Sinh 100.000 phần tử theo phân phối chuẩn N(mu, sigma^2)
# Công thức: x = mu + sigma * np.random.randn(N)
x = mu + sigma * np.random.randn(N_points)

# 3. Sử dụng np.histogram để thống kê mật độ xác suất
# bins = 300 theo yêu cầu trong phần chú thích của đề bài
cnt, edge = np.histogram(x, bins=300, density=True)

# 4. Tính toán giá trị đại diện cho mỗi cột (giữa hai cạnh của bin) để vẽ scatter
val = (edge[:-1] + edge[1:]) / 2
# Sử dụng hàm np.random.randn() sinh phân phối chuẩn N(0,1) --> N(m,s^2) = m + N(0,1) * s
# Sử dụng hàm np.histogram với density=True để thống kê số lần xuất hiện với số bins = 300 (do dãy số thực)
# ...

print( "c. Sinh day so theo phan phoi chuan")
print(f"+ x: len = {len(x)}, min [{np.min(x)}], max [{np.max(x)}]")
print(f"+ Mat so xac suat cua day so x")
plt.figure(figsize=(6,6))      # kich thuoc ban ve
plt.xlim(np.min(val), np.max(val))
plt.scatter(val, cnt)
plt.show()
```

#### 4. Giải hệ phương trình tuyến tính
Giải hệ phương trình tuyến tính sau:
$$
\left\{\begin{matrix}
        4x_1  + 3x_2 - 5x_3 = 2 \\ 
        -2x_1 - 4x_2 + 5x_3 = 5 \\ 
        8x_1 + 8x_2 = -3 \\ 
\end{matrix}\right.
$$

```python
# 1. Khai báo ma trận hệ số A và vector kết quả y
A = np.array([[4, 3, -5], 
              [-2, -4, 5], 
              [8, 8, 0]])
y = np.array([2, 5, -3])

# 2. Giải hệ phương trình Ax = y
x = np.linalg.solve(A, y)

# 3. Kiểm tra lại kết quả: yy = Ax
yy = A @ x

# 4. Tính sai số (norm bậc 2) giữa yy và y
diff = np.linalg.norm(yy - y)
# ...

print(f"+ Ma tran A: \n{A}")
print(f"+ Vector y: {y}")
print(f"+ Nghiem x: {x}")
print(f"+ Kiem tra: yy = Ax = {yy} ==> ||yy -y ||_2 = {diff: .2f}")
```

### Bài 3. Tính gần đúng số pi dùng phương pháp Monte Carlo

Vẽ một đường tròn $C$ có bán kính $r$ nội tiếp một hình vuông $C$, và đặt ngẫu nhiên $N$ chấm lên hình vuông. Tỉ lệ các chấm nằm trong hình tròn trên tổng số chấm xấp xỉ bằng diện tích của hình tròn chia cho hình vuông như sau:
$$
\frac{N_C}{N} \approx  \frac{\mathbb{S}_C}{\mathbb{S}_N}
$$
nơi mà $N_C$ là các chấm trong hình tròn, $N$ là tổng số chấm, $\mathbb{S}_C$ và $\mathbb{S}_N$ lần lượt là diện tích của hình tròn và hình vuông.
$$ \mathbb{S}_C = \pi * r^2 $$
$$ \mathbb{S}_N = (2*r)^2 = 4r^2 $$
Do đó, công thức trên trở thành:
$$
\frac{N_C}{N} \approx \frac{\pi * r^2}{4r^2} = \frac{\pi}{4}
$$
Suy ra, ta có giá trị số $\pi$ được tính như sau:
$$
\pi \approx \frac{4N_C}{N}
$$

Các bạn hãy lập trình tính sấp xỉ số $\pi$ dùng phương pháp Monte Carlo. Cho biết các sai số với $N = 100$, $N = 10000$, và $N = 1000000$. Vẽ đồ thị minh họa cho chương trình.

**Hướng dẫn**: Viết chương trình phát sinh $N$ điểm $(x_i, y_i)$ với $x_i, y_i \in [-1, -1]$. $(x_i, y_i)$ thuộc hình tròn tâm $(0, 0)$ bán kính 1 nội tiếp hình vuông có tâm tại $(0, 0)$ độ dài cạnh 2 khi và chỉ khi $x_i^2 + y_i^2 <= 1$.

```python
import matplotlib.pyplot as plt # khai báo thư viện vẽ pyplot
import math

circle = plt.Circle((0, 0), 1, color='g') # tạo đối tượng vòng tròn

fig, ax = plt.subplots(figsize=(6, 6)) # lấy figure và vùng vẽ

plt.plot(0, 0, 'o', color=(0.9, 0.9, 1.0), alpha=0.8) # vẽ điểm tại tâm (0,0)
ax.add_patch(circle) # thêm vòng tròn vào vùng vẽ

x = 0.75; y = math.sqrt(1 - (x ** 2)) # (x = 0.75, y = căn bậc 2 của 1 - x^2)
plt.arrow(0, 0, x, y)

plt.xlim(-1, 1) # giới hạn trục x
plt.ylim(-1, 1) # giới hạn trục y

plt.savefig('pi.png') # lưu đồ thị xuống tập tin
plt.show()            # hiển thị đồ thị
```

```python
import numpy as np

def calc_pi_monte_carlo(n = 100):
    pi = 0
    
    # gợi ý: sử dụng np.random.rand(s1, s2, ...) --> sinh ngẫu nhiên các số trong nửa đoạn [0.0, 1.0)
    pts = np.random.rand(n, 2) * 2 - 1
    
    # Điều kiện điểm nằm trong hình tròn: x^2 + y^2 <= 1
    inside_circle = np.sum(pts**2, axis=1) <= 1
    
    # Số điểm nằm trong (Nc)
    nc = np.sum(inside_circle)
    
    # Công thức: pi ~ 4 * Nc / N
    pi = 4 * nc / n
  
    # ...
    
    return pi
# calc_pi_monte_carlo
```

```python
import math

print("epsilon(n=100): ", calc_pi_monte_carlo(n = 100) - math.pi)
print("epsilon(n=100): ", calc_pi_monte_carlo(n = 100) - math.pi)
print("epsilon(n=10000): ", calc_pi_monte_carlo(n = 10000) - math.pi)
print("epsilon(n=10000): ", calc_pi_monte_carlo(n = 10000) - math.pi)
print("epsilon(n=1000000): ", calc_pi_monte_carlo(n = 1000000) - math.pi)
print("epsilon(n=1000000): ", calc_pi_monte_carlo(n = 1000000) - math.pi)
```
