# BÀI 1: LIỆT KÊ HOÁN VỊ (HOANVI)

Cho tập A gồm N phần tử khác nhau.  
Mỗi phần tử là một ký tự từ A → Z.

Yêu cầu:
- Liệt kê tất cả các hoán vị của tập A.

Kết quả:
- Dòng đầu: số lượng hoán vị K
- K dòng tiếp theo: mỗi dòng là một hoán vị.

```python
n = int(input())

A = []
for _ in range(n):
    A.append(input().strip())
```

```python
perm = [""] * n
used = [False] * n
result = []
```

```python
def Try(k):

    for i in range(n):

        if not used[i]:

            perm[k] = A[i]
            used[i] = True

            if k == n-1:
                result.append(perm.copy())
            else:
                Try(k+1)

            used[i] = False
```

```python
Try(0)
```

```python
print(len(result))

for p in result:
    print(*p)
```
