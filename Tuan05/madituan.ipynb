# BÀI 3: MÃ ĐI TUẦN (Knight's Tour)

Cho bàn cờ kích thước NxN.  
Con mã bắt đầu tại vị trí (x0, y0).

Yêu cầu:
- Đi qua tất cả các ô đúng một lần
- Cuối cùng quay lại ô ban đầu bằng một nước mã

Thuật toán sử dụng: **Backtracking**

```python
n, x0, y0 = map(int, input().split())
```

```python
board = [[0]*(n+1) for _ in range(n+1)]

dx = [-2,-2,-1,-1,1,1,2,2]
dy = [-1,1,-2,2,-2,2,-1,1]

board[x0][y0] = 1
found = False
```

```python
def printBoard():

    print(1)

    for i in range(1,n+1):
        for j in range(1,n+1):
            print(board[i][j], end=" ")
        print()
```

```python
def valid(x,y):

    return 1 <= x <= n and 1 <= y <= n and board[x][y] == 0
```

```python
def Try(x,y,step):

    global found

    if found:
        return

    if step == n*n:

        for i in range(8):
            nx = x + dx[i]
            ny = y + dy[i]

            if nx == x0 and ny == y0:
                printBoard()
                found = True
                return
        return

    for i in range(8):

        nx = x + dx[i]
        ny = y + dy[i]

        if valid(nx,ny):

            board[nx][ny] = step + 1

            Try(nx,ny,step+1)

            board[nx][ny] = 0
```

```python
Try(x0,y0,1)

if not found:
    print(0)
```
