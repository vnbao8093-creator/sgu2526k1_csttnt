# Liệt kê Tổ Hợp (Backtracking)

Bài toán:
Cho tập hợp A có n phần tử và số k.  
Hãy liệt kê tất cả các tổ hợp chập k của tập A.

Input file: `TOHOP.INP`

Dòng 1: n k  
Các dòng tiếp theo: các phần tử của tập A

Output:
- In số lượng tổ hợp
- In tất cả tổ hợp

```python
def docTapTin(fpath = "TOHOP.INP"):
    with open(fpath, "rt") as file:
        content = file.readlines()
        
        n, k = [int(si.replace('\n','')) for si in content[0].split(' ')[:2]]
        th = [si[0] for si in content[1:]]
        
    return n, k, th
```

```python
def InDapAn(tohop, th):
    for i in range(1, len(tohop)):
        print(th[tohop[i]-1], end=' ')
    print()
```

```python
def LietKeToHop(n, k, th):

    tohop = [-1]*(k+1)
    tohop[0] = 0

    info = dict(tohop = tohop, cnt = 0, n = n, k = k, th = th, isprint = False)

    TryToHop(1, info)

    print(info['cnt'])


    tohop = [-1]*(k+1)
    tohop[0] = 0

    info = dict(tohop = tohop, cnt = 0, n = n, k = k, th = th, isprint = True)

    TryToHop(1, info)
```

```python
def TryToHop(t, info):

    tohop = info["tohop"]
    k = info["k"]
    n = info["n"]

    for vt in range(tohop[t-1]+1, n - k + t + 1):

        tohop[t] = vt

        if t == k:

            info["cnt"] = info["cnt"] + 1

            if info['isprint']:
                InDapAn(tohop, info["th"])

        else:

            TryToHop(t+1, info)

        tohop[t] = -1
```

```python
n, k, th = docTapTin("TOHOP.INP")

print("n =", n)
print("k =", k)
print("tap hop =", th)

print("\nKet qua:")
LietKeToHop(n, k, th)
```
