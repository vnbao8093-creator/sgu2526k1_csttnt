# BÀI 2: ĐIỀN BẢNG SỐ (FUJIWARA)

Cho bảng Sudoku kích thước 9×9.

Yêu cầu:
- Điền các số từ 1 → 9 vào các ô trống
- Mỗi hàng chứa đủ 1 → 9
- Mỗi cột chứa đủ 1 → 9
- Mỗi ô 3×3 chứa đủ 1 → 9

Input:
- 9 dòng
- '.' là ô trống

Output:
- bảng Sudoku hoàn chỉnh
- hoặc in "IMPOSSIBLE" nếu không có lời giải

```python
board = []

for _ in range(9):
    row = list(input().strip())
    board.append(row)
```

```python
def valid(r,c,num):

    num = str(num)

    for i in range(9):
        if board[r][i] == num:
            return False

    for i in range(9):
        if board[i][c] == num:
            return False

    sr = (r//3)*3
    sc = (c//3)*3

    for i in range(3):
        for j in range(3):
            if board[sr+i][sc+j] == num:
                return False

    return True
```

```python
def findEmpty():

    for i in range(9):
        for j in range(9):
            if board[i][j] == '.':
                return i,j

    return None
```

```python
def solve():

    pos = findEmpty()

    if not pos:
        return True

    r,c = pos

    for num in range(1,10):

        if valid(r,c,num):

            board[r][c] = str(num)

            if solve():
                return True

            board[r][c] = '.'

    return False
```

```python
if solve():

    for row in board:
        print("".join(row))

else:
    print("IMPOSSIBLE")
```
