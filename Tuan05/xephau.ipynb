# Bài toán Xếp Hậu (N-Queens)

Cho bàn cờ kích thước N×N.

Yêu cầu:
Đặt N quân hậu lên bàn cờ sao cho:
- Không có hai quân hậu nào cùng hàng
- Không có hai quân hậu nào cùng cột
- Không có hai quân hậu nào cùng đường chéo

Input:
File `XEPHAU.INP` chứa số nguyên N

Output:
- In số lời giải
- In ma trận bàn cờ với:
    1 : vị trí quân hậu
    0 : ô trống

Thuật toán sử dụng: **Backtracking**

```python
def docTapTin(fpath="XEPHAU.INP"):
    with open(fpath, "rt") as file:
        content = file.read()
        n = int(content)
    return n
```

```python
def InDapAn(info):

    print(info["cnt"])

    for i in range(1, info["n"]+1):
        for j in range(1, info["n"]+1):

            if info["hau"][i] == j:
                print("1", end=" ")
            else:
                print("0", end=" ")

        print()

    print()
```

```python
def TryHau(k, info):

    if info["coDapAn"] and info["action"] == "find":
        return

    n = info["n"]
    hau = info["hau"]
    dong = info["dong"]
    cheo1 = info["cheo1"]
    cheo2 = info["cheo2"]

    for vk in range(1, n+1):

        if dong[vk]==0 and cheo1[vk-k+n]==0 and cheo2[vk+k-1]==0:

            hau[k] = vk
            dong[vk] = 1
            cheo1[vk-k+n] = 1
            cheo2[vk+k-1] = 1

            if k == n:

                info["coDapAn"] = True
                info["cnt"] += 1

                if info["laInDapAn"]:
                    InDapAn(info)

            else:

                TryHau(k+1, info)

            hau[k] = -1
            dong[vk] = 0
            cheo1[vk-k+n] = 0
            cheo2[vk+k-1] = 0
```

```python
def LietKeHau(n, action="list"):

    hau = [-1]*(n+1)
    dong = [0]*(n+1)
    cheo1 = [0]*(2*n)
    cheo2 = [0]*(2*n)

    info = {
        "n": n,
        "hau": hau,
        "dong": dong,
        "cheo1": cheo1,
        "cheo2": cheo2,
        "coDapAn": False,
        "action": action,
        "cnt": 0,
        "laInDapAn": True
    }

    TryHau(1, info)

    return info
```

```python
n = docTapTin("XEPHAU.INP")

print("n =", n)
print("\nKet qua:")

info = LietKeHau(n)

if info["coDapAn"] == False:
    print("0")
```
